### Load Train, Test and Validation Sets

```python
import pandas as pd

# Read the datasets from the CSV files
X_resampled = pd.read_csv('X_resampled_train.csv')
y_resampled = pd.read_csv('y_resampled_train.csv').iloc[:, 0]

X_val = pd.read_csv('X_val.csv')
y_val = pd.read_csv('y_val.csv').iloc[:, 0]

X_test = pd.read_csv('X_test.csv')
y_test = pd.read_csv('y_test.csv').iloc[:, 0]
```

### SVM Implementation (with oversampling)

#### Hyperparameter Tuning and Validation

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import RandomizedSearchCV
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters to be searched
param_grid = {
    'C': [0.1, 1, 10, 100],  # Regularization parameter
    'gamma': [1, 0.1, 0.01, 0.001],  # Kernel coefficient for 'rbf', 'poly' and 'sigmoid'
    'kernel': ['linear', 'rbf']  # Specify the type of SVM
}

# Number of iterations for randomized search
n_iter_search = 10

# Create a RandomizedSearchCV object
random_search = RandomizedSearchCV(SVC(random_state=42), param_distributions=param_grid,
                                   n_iter=n_iter_search, cv=3, verbose=2, n_jobs=-1)

# Fit the model with the parameter candidates
random_search.fit(X_resampled, y_resampled)

# Print best parameters
print("Best Parameters: ", random_search.best_params_)

# Use the best model to predict on the validation set
y_val_pred = random_search.predict(X_val)

# Evaluate the model on the validation set
print(f'Accuracy on validation set with best parameters: {accuracy_score(y_val, y_val_pred)}\n')

# Display the classification report for the validation set
print(classification_report(y_val, y_val_pred))

# Visualize the confusion matrix using seaborn
conf_matrix = confusion_matrix(y_val, y_val_pred)
sns.heatmap(conf_matrix, annot=True, fmt="d")
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

#### Apply on Test Set

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Create the SVM model with a rbf kernel
svm_model = SVC(kernel='rbf', gamma=0.001, C=10, random_state=42)

# Fit the model to the resampled training data
svm_model.fit(X_resampled, y_resampled)

# Make predictions on the test data
y_test_pred = svm_model.predict(X_test)

# Evaluate the model on the test set
print(f'Accuracy on test set: {accuracy_score(y_test, y_test_pred)}\n')
```

#### Evaluation

```python
from sklearn.metrics import (accuracy_score, balanced_accuracy_score, average_precision_score, f1_score, 
                             precision_score, recall_score, cohen_kappa_score, roc_curve, auc)
import pandas as pd
import matplotlib.pyplot as plt

# 1. Accuracy and Balanced Accuracy
print(f'Accuracy: {accuracy_score(y_test, y_test_pred)}')
print(f'Balanced Accuracy: {balanced_accuracy_score(y_test, y_test_pred)}')

# 2. Average Precision - Multi-class
y_test_bin = pd.get_dummies(y_test)  # Convert to one-hot encoding
y_proba_ovr = svm_model.decision_function(X_test)  # Decision function values for multi-class
print(f'Average Precision: {average_precision_score(y_test_bin, y_proba_ovr, average="macro")}')

# 3. F1 Score, Precision, Recall
print(f'F1 Score (Micro): {f1_score(y_test, y_test_pred, average="micro")}')
print(f'F1 Score (Macro): {f1_score(y_test, y_test_pred, average="macro")}')
print(f'F1 Score (Weighted): {f1_score(y_test, y_test_pred, average="weighted")}')
print(f'Precision: {precision_score(y_test, y_test_pred, average="weighted")}')
print(f'Recall: {recall_score(y_test, y_test_pred, average="weighted")}')

# 4. Cohen's Kappa
print(f"Cohen's Kappa: {cohen_kappa_score(y_test, y_test_pred)}")

# 5. Classification Report
print('Classification Report:')
print(classification_report(y_test, y_test_pred))

# 6. ROC Curve for multi-class using one-vs-rest approach
categories_order = ['Good', 'Satisfactory', 'Moderate', 'Poor', 'Very Poor', 'Severe']
n_classes = len(categories_order)

fpr = {}
tpr = {}
roc_auc = {}

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin.iloc[:, i], y_proba_ovr[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot the ROC curves
plt.figure(figsize=(10, 8))
colors = ['blue', 'red', 'green', 'yellow', 'purple', 'cyan']
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label='{0} (area = {1:0.2f})'.format(categories_order[i], roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) for Multi-class')
plt.legend(loc="lower right")
plt.show()

# Get confusion matrix
cm = confusion_matrix(y_test, y_test_pred)

# Define the class labels
class_labels = ['Good', 'Satisfactory', 'Moderate', 'Poor', 'Very Poor', 'Severe']

# Create a DataFrame from the confusion matrix
cm_df = pd.DataFrame(cm, index=class_labels, columns=class_labels)

# Plot the confusion matrix
plt.figure(figsize=(10, 6))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

### Evaluation and Results

<b> 1. Overview: </b>

The results presented are for the air quality classification task after applying the Synthetic Minority Over-sampling Technique (SMOTE) to balance the dataset. The model used for the classification was the Support Vector Machine (SVM) with a radial basis function (RBF) kernel. The hyperparameters used for the model are gamma=0.001, C=10. The model's validation accuracy is approximately 0.7537, which is almost similar to the test accuracy.

<b> 2. Key Metrics & Analysis: </b>

1. Accuracy: The test accuracy stands at approximately 75.25%. Given that this is in close alignment with the validation accuracy of 75.37%, this suggests that the model generalizes well to unseen data.

2. Balanced Accuracy: A score of 74.17% suggests that the model has an almost equal performance rate for both the positive and negative classes in binary classification. This is a good indicator, especially for imbalanced datasets.

3. Average Precision: With a score of 72.03%, the model's capability to retrieve only relevant instances (positive true rate) over the total retrieved instances is commendable.

4. F1 Score (Micro, Macro, and Weighted): The micro F1 score, macro F1 score, and weighted F1 score are 75.25%, 72.04%, and 75.44% respectively. These metrics take both false positives and false negatives into account. The close alignment in these scores indicates a balanced performance across classes.

5. Precision: The models ability to correctly predict the positive class from all the predicted positives stands at approximately 75.86%.

6. Recall: The test recall is approximately 75.25%, indicating the model's ability to correctly identify the positive class from all the actual positives.

7. Cohen's Kappa: A score of 0.668 signifies a good level of agreement between the predicted and actual values, which is higher than what would be expected by chance.

8. Classification Report: Analyzing class-wise performance:
    - Class Good and Satisfactory have relatively high precision, recall, and F1-score values, indicating strong performance.
    - Class Moderate has good performance but slightly lesser compared to the above two.
    - Classes Poor, Very Poor, and Severe exhibit variability in their scores, indicating that there might be some challenges in distinguishing these categories.

9. ROC Curve: The Receiver Operating Characteristic (ROC) curve showcases the trade-off between sensitivity and specificity for different threshold values. The areas under the curve (AUC) for each class are above 0.90, which is excellent, indicating strong model performance.

10. Confusion Matrix: The matrix visually represents the true positives, true negatives, false positives, and false negatives. For instance, the Moderate class has the highest confusion with other classes, whereas Satisfactory has a high true positive rate.

<b> 3. Conclusions: </b>

- The model exhibits a strong and consistent performance across multiple metrics.
- The close alignment of validation and test accuracies suggests good generalization to new data.
- While the model performs well for certain classes, there is room for improvement in distinguishing some of the air quality classes, as seen from the confusion matrix.
- Given the performance metrics and the ROC curve, the model can be considered robust and reliable for this air quality classification task.
